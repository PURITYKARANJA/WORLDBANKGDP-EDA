```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="Year",y="GDP (USD)",hue="Country",palette="viridis",data=df)
plt.title("Year vs GDP(USD)")
plt.xlabel("Year")
plt.ylabel("GDP (USD)")
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="Population",y="GDP (USD)",hue="Country",palette="tab20",data=df)
plt.title("Poplation vs GDP (USD)")
plt.xlabel("Population")
plt.ylabel("GDP (USD)")
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x="Year",y="Population",hue="Country",palette="viridis",data=df)
plt.title("Measure of Population Growth Over Time")
plt.xlabel("Year")
plt.ylabel("Population")
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1.05,1),loc="upper left")
plt.show()
```

```python
fig,axes=plt.subplots(2,2,figsize=(10,6))
plt.suptitle("Measures of Distribution")
sns.histplot(df['Life Expectancy'],kde=True,ax=axes[0,0]).set_title("Distribution of Life Expectancy")
sns.histplot(df['Unemployment Rate (%)'],kde=True,ax=axes[0,1],color="green").set_title("Distribution of Unemployment Rate (%)")
sns.histplot(df['CO2 Emissions (metric tons per capita)'],kde=True,ax=axes[1,0],color="orange").set_title("Distribution of CO2 Emissions (metric tons per capita)")
sns.histplot(df['Access to Electricity (%)'],kde=True,ax=axes[1,1],color="skyblue").set_title("Distribution ofAccess to Electricity (%)")
plt.tight_layout()
plt.show()
```

```python
data=df.select_dtypes(include=np.number)
corr=data.corr()
plt.figure(figsize=(10,6))
sns.heatmap(corr,annot=True,cmap="viridis")
plt.title("GPD CAUSALITY CORRELATION MATRIX")
plt.show()
```

```python
import plotly.express as px
data_group=df.groupby('Country')[["GDP (USD)","Unemployment Rate (%)"]].mean().reset_index()
fig=px.scatter(
    data_group,
    title=("Unemployment effects on GDP As Per Country"),
    x="Unemployment Rate (%)",
    y="GDP (USD)",
    text="Country",
    labels={"Unemployment Rate (%)":"Average Unemployment Rate (%)","GDP (USD)":"Average GDP (USD)"},
    hover_name="Country",
)
fig.show()
```

```python
data=df.groupby("Country",as_index=False).agg({"GDP (USD)":"mean","Population":"mean","Life Expectancy":"mean"})
fig=px.scatter_3d(
    data,
    x="GDP (USD)",
    y="Population",
    z="Life Expectancy",
    text="Country",
    title="GDP, Population, and Life Expectancy Across Countries",
    labels={"GDP (USD)":"Average GDP (USD)","Population":"Average Population","Life Expectancy":"Average Life Expectancy"},
    color="Country",
    template="plotly_dark",
)
fig.show()
```

```python
df_new=df.groupby("Country",as_index=False).agg({"GDP (USD)":"mean","CO2 Emissions (metric tons per capita)":"mean"})
fig=px.scatter(
    df_new,
    x="GDP (USD)",
    y="CO2 Emissions (metric tons per capita)",
    size="CO2 Emissions (metric tons per capita)",
    size_max=60,
    text="Country",
    title="GDP VS CO2 EMISSIONS",
    labels={"GDP (USD)":"Average GDP (USD)","CO2 Emissions (metric tons per capita)":"Average CO2 Emissions (metric tons per capita)"},
    hover_name="Country",
    color="Country",
    template="plotly_dark",
)
fig.show()
```

```python
fig=px.scatter(
    df,
    x="GDP (USD)",
    y="Population",
    size="Life Expectancy",
    animation_frame="Year",
    size_max=55,
    color="Unemployment Rate (%)",
    hover_name="Country",
    title=("GDP VS POPULATION VS LIFE EXPECTANCY VS UNEMPLOYMENT RATE"),
    template="plotly_dark",
)
fig.show()
```

```python
fig=px.scatter(
    df,
    x="GDP (USD)",
    y="Population",
    size="Life Expectancy",
    animation_frame="Year",
    size_max=55,
    color="Access to Electricity (%)",
    hover_name="Country",
    title=("GDP VS POPULATION VS LIFE EXPECTANCY VS Access to Electricity (%)"),
    template="plotly_dark",
)
fig.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Country', y='GDP (USD)', palette='viridis')
plt.title('GDP by Country')
plt.ylabel('GDP (USD)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
fig = px.scatter_geo(
    df,
    locations='Country',
    locationmode='country names',
    size='Access to Electricity (%)',
    color='Access to Electricity (%)',
    hover_name='Country',
    animation_frame='Year',
    projection='natural earth',
    title='Access to Electricity by Country Over Time',
    size_max=10,
    color_continuous_scale='RdYlGn',
    template='plotly_dark'
)

fig.show()
```
